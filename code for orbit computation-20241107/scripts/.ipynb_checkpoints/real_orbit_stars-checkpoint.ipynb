```python

import numpy as np
import pylab as pl
import astropy.units as u
from galpy.orbit import Orbit
import galpy.util.coords as coords
from galpy.potential import MWPotential2014

```

Loading the CSV file into a python array. The CSV file is the Gaia 100-pc white dwarf sample that was classified by [Torres et al. 2019, MNRAS, 485, 5573-5589](https://ui.adsabs.harvard.edu/abs/2019MNRAS.485.5573T/abstract)) with a random forest algorithm.

```python
#np.random.seed(0)

# for local files, just use this function #
S = np.genfromtxt('../data/Gaia100pcStars.csv', delimiter=',', names=True, dtype=None, encoding="ASCII")
print(S.dtype)


```

We create orbits for non white dwarf stars, selected from Gaia within a 100-pc volume, respectively. These objects have good quality radial velocity measurements.


```python
R0, z0, V0, v_sun = 8.122*u.kpc, 0.0208*u.kpc, 235.*u.km/u.s, [12.9, 10.6, 7.78]*u.km / u.s



ts = np.linspace(0,0.125,1000)*u.Gyr

a = Orbit([S['ra'][0:50000],S['dec'][0:50000],1/S['parallax'][0:50000],
                S['pmra'][0:50000], S['pmdec'][0:50000], S['radial_velocity'][0:50000]],
                radec=True, ro=R0, zo=z0, vo=V0, solarmotion=v_sun)

a.integrate(ts,MWPotential2014)
```

Plotting the Toomre diagram of the Full sample.

```python

fig = pl.figure(figsize=(7,7))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(111)

ax1.plot(a.V(),np.sqrt(a.U()**2 + a.W()**2), marker='.', ms=3, ls='none', label='Normal Stars')

ax1.set_xlabel(r'$V_{\rm LSR}$ [km/s]')
ax1.set_ylabel(r'$\sqrt{U^2_{\rm LSR} + W^2_{\rm LSR}}$ [km/s]')

pl.legend(frameon=False)


pl.show()
```

Plotting the vertical component of angular momentum (Lz) vs orbital eccentricity.

```python
fig = pl.figure(figsize=(7,7))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(111)

ax1.plot(a.e(),a.Lz(), marker='.', ms=0.5, ls='none', label='Normal Stars')

ax1.set_xlabel(r'$e$')
ax1.set_ylabel(r'$L_z$ [kpc\,km/s$^2$]')

pl.legend(frameon=False)


pl.show()
```

Plotting the orbits

```python
fig = pl.figure(figsize=(14,6))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(121)

ax1.plot(a.R(ts,quantity=True)[0],a.z(ts,quantity=True)[0])



ax1.set_xlabel('R [kpc]')
ax1.set_ylabel('Z [kpc]')


ax2 = fig.add_subplot(122)

ax2.plot(a.y(ts,quantity=True)[0],a.x(ts,quantity=True)[0], label='halo')

ax2.set_xlabel('Y [kpc]')
ax2.set_ylabel('X [kpc]')

pl.legend(frameon=False)

ax2.invert_xaxis()
ax2.invert_yaxis()
pl.show()
```

Integrating for longer and plotting the orbits

```python
ts = np.linspace(0,5,10000)*u.Gyr


a100 = Orbit([S['ra'][100],S['dec'][100],1/S['parallax'][100],
                S['pmra'][100], S['pmdec'][100], S['radial_velocity'][100]],
                radec=True, ro=R0, zo=z0, vo=V0, solarmotion=v_sun)

a100.integrate(ts,MWPotential2014)
print(a100.R(ts))

fig = pl.figure(figsize=(14,6))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(121)

ax1.plot(a100.R(ts),a100.z(ts,quantity=True))

ax1.set_xlabel('R [kpc]')
ax1.set_ylabel('Z [kpc]')


ax2 = fig.add_subplot(122)

ax2.plot(a100.y(ts,quantity=True),a100.x(ts,quantity=True), label='halo')

ax2.set_xlabel('Y [kpc]')
ax2.set_ylabel('X [kpc]')

pl.legend(frameon=False)

ax2.invert_yaxis()
pl.show()


```

Plotting the Galactic radial velocity component:

```python
fig = pl.figure(figsize=(7,7))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(111)

ax1.plot(a100.R(ts,quantity=True),a100.vR(ts,quantity=True), label='halo')


pl.legend(frameon=False)

ax1.set_xlabel('R [kpc]')
ax1.set_ylabel('vR [km/s]')
pl.show()
```
