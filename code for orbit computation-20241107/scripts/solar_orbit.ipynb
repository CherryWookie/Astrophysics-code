Importing the Orbit class and MWPotential class, which is representative of our Galaxy's potential (see next lesson).

```python
import numpy as np
import pylab as pl
import astropy.units as u
from galpy.orbit import Orbit
from galpy.potential import MWPotential2014
```

We create an empty Orbit, which is meant for the Sun.
Then we create an array of timesteps
And finally we integrate the orbit.

```python
o = Orbit()

o = Orbit([o.R(quantity=True),o.vR(quantity=True),
              o.vT(quantity=True),
              o.z(quantity=True),o.vz(quantity=True),
              o.phi(quantity=True)])

ts = np.linspace(0,0.125,10000)*u.Gyr
f = o.flip()

o.integrate(ts,MWPotential2014)
f.integrate(ts,MWPotential2014)

print('e = ', round(o.e(),3), 'Lz = ', o.Lz(), 'kpc * km/s')
```

Plotting the orbit

```python
fig = pl.figure(figsize=(7,3))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(121)
ax1.plot(f.R(ts),f.z(ts), label='past orbit')
ax1.plot(o.R(ts),o.z(ts), label='future orbit')
ax1.set_xlabel('R [kpc]')
ax1.set_ylabel('Z [kpc]')


ax2 = fig.add_subplot(122)

ax2.plot(f.y(ts),f.x(ts), label='past orbit')
ax2.plot(o.y(ts),o.x(ts), label='future orbit')
ax2.set_xlabel('Y [kpc]')
ax2.set_ylabel('X [kpc]')

pl.legend(frameon=False)

ax2.invert_xaxis()
ax2.invert_yaxis()
pl.show()
```

```python
fig = pl.figure(figsize=(5,5))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(projection='3d')
ax1.plot(f.y(ts),f.x(ts), f.z(ts), label='past orbit')
ax1.plot(o.y(ts),o.x(ts), o.z(ts), label='future orbit')
ax1.set_xlabel('Y [kpc]')
ax1.set_ylabel('X [kpc]')
ax1.set_zlabel('Z [kpc]')

ax1.invert_xaxis()
ax1.invert_yaxis()
pl.show()
```

Integrating for longer and plotting the orbit

```python
ts = np.linspace(0,4.5,10000)*u.Gyr
f = o.flip()

o.integrate(ts,MWPotential2014)
f.integrate(ts,MWPotential2014)

print('e = ', round(o.e(),3), 'Lz = ', o.Lz(), 'kpc * km/s')

fig = pl.figure(figsize=(7,3))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(121)
ax1.plot(f.R(ts),f.z(ts), label='past orbit')
ax1.plot(o.R(ts),o.z(ts), label='future orbit')
ax1.set_xlabel('R [kpc]')
ax1.set_ylabel('Z [kpc]')



ax2 = fig.add_subplot(122)

ax2.plot(f.y(ts),f.x(ts), label='past orbit')
ax2.plot(o.y(ts),o.x(ts), label='future orbit')
ax2.set_xlabel('Y [kpc]')
ax2.set_ylabel('X [kpc]')

pl.legend(frameon=False)

ax2.invert_yaxis()
pl.show()


```

Plotting radial velocity:

```python
fig = pl.figure(figsize=(5,5))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(111)
ax1.plot(o.R(ts),o.vR(ts), label='future orbit')
ax1.set_xlabel('R [kpc]')
ax1.set_ylabel('vR [km/s]')
pl.show()
```
